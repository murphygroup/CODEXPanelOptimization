```python
import numpy as np

"""
For the initial matrix, we want to put overlap at the center of the matrix. 
So, we need some preprocesses for our matrix from each single panel. 
For panelA's matrix, we want to move the overlap part to the end;
for panel B's matrix, we want to move the overlap part to the head.
"""
def rearrange(idxes, idx, mtr):
    idxes2 = idxes.copy()
    idxes2.remove(idx)
    idxes2.insert(len(idxes)-1, idx)
    new_mtr = np.zeros((len(idxes),len(idxes)))

    for i in range(len(idxes)):
        for j in range(len(idxes)):
            ii = idxes[i]
            jj = idxes[j]      
            new_mtr[idxes2.index(ii),idxes2.index(jj)] = mtr[i,j]    
    return new_mtr, idxes2  

def rearrange2(idxes, idx, mtr):
    idxes2 = idxes.copy()
    idxes2.remove(idx)
    idxes2.insert(0, idx)
    new_mtr = np.zeros((len(idxes),len(idxes)))

    for i in range(len(idxes)):
        for j in range(len(idxes)):
            ii = idxes[i]
            jj = idxes[j]      
            new_mtr[idxes2.index(ii),idxes2.index(jj)] = mtr[i,j]    
    return new_mtr, idxes2 
```

```python
overlap = [34, 9, 43, 35, 14, 0]
p1 = [0, 1, 2, 3, 4, 9, 10, 12, 14, 15, 16, 17, 20, 22, 25, 26, 33, 34, 35, 37, 38, 39, 41, 42, 43, 45] # panelA marker indexes
p2 = [0, 5, 6, 7, 8, 9, 11, 13, 14, 18, 19, 21, 23, 24, 27, 28, 29, 30, 31, 32, 34, 35, 36, 40, 43, 44] # panelB marker indexes

m1 = np.load('mae_mtx_setA_zscore.npy')
m2 = np.load('mae_mtx_setB_zscore.npy')

# rearrange the orderm1 and m2
idxes = p1.copy()
m11 = m1.copy()
for i in overlap:
    m11, idxes = rearrange(idxes, i, m11)
print(idxes)

idxes = p2.copy()
m22 = m2.copy()
for i in overlap[::-1]:
    m22, idxes = rearrange2(idxes, i, m22)
print(idxes)

m = np.zeros((46,46))
```

```python
m[:26, :26] += m11
m[20:, 20:] += m22
m[20:26, 20:26] /= 2

init_m = m.copy()

for i in range(20):
    for j in range(26, 46):
        m[i, j] = np.min(m[20:26, i] + m[20:26, j])

m[26:,:20] = m[:20, 26:].copy().transpose()
```
