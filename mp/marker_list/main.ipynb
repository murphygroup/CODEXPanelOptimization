```python
import os
import pandas as pd
import numpy as np
```

```python
path = os.getcwd()

# find all the txt files
files = []
for r, d, f in os.walk(path):
    for file in f:
        if '.txt' in file:
            files.append(os.path.join(r, file))
files = sorted(files, key=lambda x: (x.split('/')[-1].split('.')[0][:2], int(x.split('/')[-1].split('.')[0][2:])))

# read all the txt files    
marker_list = []
for f in files:
    with open(f, 'r') as file:
        marker_list.append(file.read().splitlines())
```

```python
all_markers = []
for markers in marker_list:
    for marker in markers:
        marker = marker.upper()
        if marker == 'HLA-DR':
            marker = 'HLADR'
        elif marker == 'DAPI-02' or marker == 'DAPI01':
            marker = 'DAPI'
        elif marker == 'COLL4' or marker == 'COLLAGENIV':
            marker = 'COLLLV'
        elif marker == 'PODOPIANIN':
            marker = 'PODOPLANIN'
        elif marker == 'SYNAPTO':
            marker = 'SYNPO'
        elif marker =='HOECHST1':
            marker = 'HOECHST'
        if marker not in all_markers:   
            all_markers.append(marker)

all_markers.sort()
```

```python
all_markers
```

```python
# create a dataframe where each row is a marker, each column is a dataset(each txt file), and each entry is whether this dataset has this marker or not
files_name = []
for f in files:
    files_name.append(f.split('.')[0].split('/')[-1])
df = pd.DataFrame(columns = files_name, index = all_markers)
```

```python
for i, markers in enumerate(marker_list):
    for marker in markers:
        marker = marker.upper()
        if marker == 'HLA-DR':
            marker = 'HLADR'
        elif marker == 'DAPI-02' or marker == 'DAPI01':
            marker = 'DAPI'
        elif marker == 'COLL4' or marker == 'COLLAGENIV':
            marker = 'COLLLV'
        elif marker == 'PODOPIANIN':
            marker = 'PODOPLANIN'
        elif marker == 'SYNAPTO':
            marker = 'SYNPO'
        elif marker =='HOECHST1':
            marker = 'HOECHST',
        df.loc[marker, files_name[i]] = 1
```

```python
df
```

```python
# highlight entries that are 1
def highlight_ones(s):
    '''
    highlight the entries that are 1
    '''
    is_one = s == 1
    return ['background-color: yellow' if v else '' for v in is_one]
```

```python
for name in files_name:
    assert df[name].sum() == int(name[-2:])

df = df.iloc[:, :-1]
df.style.apply(highlight_ones, axis = 1)
```

```python
df.style.apply(highlight_ones, axis = 1).to_html('marker_list.html')
```
